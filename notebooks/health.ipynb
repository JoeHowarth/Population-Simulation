```python
def logistic(r, N, K):
    return r * N *(K - N) / K
```

```python
ys = [.9]
ts = [1.2] * 5 + [0.9] * 5 + [1.1] * 5 + [0.6] * 5

for i in range(20):
    v = ys[i] + logistic(.05, ys[i], ts[i])
    ys.append(v)
plt.plot(ys)
```
